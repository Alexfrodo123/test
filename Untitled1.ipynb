```python
import re
mac_address_table = open('Крутота.txt', encoding  = 'UTF-8').read()

slova = 'ростелеком'

pattern = r'\b[ростелкм]{1,10}\n'

prog = re.compile( pattern )

result = prog.findall(mac_address_table)

resultat = list(dict.fromkeys(result))


res = []
k = 0
r1 = 0
o1 = 0
c1 = 0
t1 = 0
e1 = 0
l1 = 0
k1 = 0
m1 = 0

for word in resultat:
    for simv in word:
        if simv == 'р':
            r1 = r1 + 1
            if r1 == 2:
                k = 1
        if simv == 'о':
            o1 = o1 +1
            if o1 == 3:
                k = 1
        if simv == 'с':
            c1 = c1 +1
            if c1 == 2:
                k = 1
        if simv == 'т':
            t1 = t1 +1
            if t1 == 2:
                k = 1
        if simv == 'е':
            e1 = e1 +1
            if o1 == 3:
                k = 1
        if simv == 'л':
            l1 = l1 +1
            if l1 == 2:
                k = 1
        if simv == 'к':
            k1 = k1 +1
            if o1 == 2:
                k = 1
        if simv == 'м':
            m1 = m1 +1
            if o1 == 2:
                k = 1
    if k == 0:
        res.append( word )
    k = 0
    r1 = 0
    o1 = 0
    c1 = 0
    t1 = 0
    e1 = 0
    l1 = 0
    k1 = 0
    m1 = 0     

        
#print('конечный результат', res) 

res2 = []

hjk = ''

for wor in res:
    for jk in wor:
        if jk != '\n':
            hjk = hjk + jk
    res2.append(hjk) 
    hjk = ''

stringRes = ', '.join( res2 )
print('Из слова ростелеком можно составить данные слова:', stringRes)
    
```
